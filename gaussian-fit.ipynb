```python
from scipy.optimize import curve_fit
import numpy as np
```

```python
def gaussian(x, a, b, c):
    return a*np.exp(-(x-b)**2/c**2)
```

```python
popt, junk = curve_fit(gaussian, data.x, data.y, p0=[1, 0, 1])
```

```python
plt.scatter(data.x, data.y)

xmodel = np.linspace(-5, 5, 1000)
ymodel = gaussian(xmodel, *popt)
plt.plot(xmodel, ymodel, lw=2, color='red')
```
